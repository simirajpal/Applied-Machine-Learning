### Approximating images with neural networks

a) Describe the structure of the network

This is a 2-layer neural network with 7 hidden layers. Each hidden layer has 20 neurons that perform weighted additions of the inputs and passes them throuh a ReLU (Rectified Linear Unit) activation function. The first layer takes the pixel's (x,y) position on the grid, and the network predicts the (r,g,b) color at that point using a regression layer as its loss layer at the end.

b) Explain what loss is

The loss quantifies the of the predicted output image from the actual output image. In this example, the loss layer is an L2 regression layer. It penalizes sum(||x - y||^2), where x is the input and y is the user-provided "correct" values.

### Random forests for image approximation

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

#### a) Get Mona Lisa Image

```python
mona = imageio.imread('mona_lisa.jpg')
```

```python
plt.imshow(mona)
```

#### b) Preprocessing of Input

```python
# check shape
mona.shape
```

```python
# randomly sample coordinates
x_rand = np.random.randint(0, 900, 5000)
y_rand = np.random.randint(0, 604, 5000)

# put them together
coord_rand = [[x_rand[i], y_rand[i]] for i in range(5000)]

# check correct shape
(np.asarray(coord_rand)).shape
```

To preprocess the data, we can normalize the image by dividing it by 255. This would allow the dot products to be 
done more efficiently and prevent the data from keeping increasing with the products. Since it is also needed for imshow to plot later, we can normalize at this point and forget about it. Other preprocessing steps include filling missing data values, or converting categorical values to numerical values, but these are not necessary in this case.

```python
# normalize it
mona = mona/255
```

#### c) Preprocessing of Output

```python
from skimage.color import rgb2gray
gray_mona = rgb2gray(mona)
```

```python
plt.imshow(gray_mona, cmap = 'gray')
plt.savefig('Mona_Grey.png')
```

Since the necessary information in the image is available in grayscale, we can just perform the model on the grayscale image. This automatically gives us normalized values for the image (i.e. pixel values 0 to 1 instead of 0 to 255).

```python
sample = [gray_mona[coord_rand[i][0]][coord_rand[i][1]] for i in range(5000)]
```

# more post processing steps?

d) Feed coordinates to random forest and color pixels with result

```python
y_training = [] # to store the colored pixels - esentially the labels
mona_coord = []

for i in range(900):
    for j in range(604):
        mona_coord.append((i,j))

for i in range(5000):
    y_training.append(list(mona[x_rand[i]][y_rand[i]]))

# set up regressor object 
regressor = RandomForestRegressor(max_depth=20) 

# fit regressor 
regressor.fit(coord_rand, y_training) 

# predict
predicted_lisa = regressor.predict(mona_coord)

# resshape
lisa_reshape = predicted_lisa.reshape(900,604, 3)

# display predicted image
plt.imshow(lisa_reshape)
plt.savefig('Lisa_regression.png')
```
