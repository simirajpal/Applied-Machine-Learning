### Approximating images with neural networks

a) Describe the structure of the network

This is a 2-layer neural network with 7 hidden layers. Each hidden layer has 20 neurons that perform weighted additions of the inputs and passes them throuh a ReLU (Rectified Linear Unit) activation function. The first layer takes the pixel's (x,y) position on the grid, and the network predicts the (r,g,b) color at that point using a regression layer as its loss layer at the end.

b) Explain what loss is

The loss quantifies the of the predicted output image from the actual output image. In this example, the loss layer is an L2 regression layer. It penalizes sum(||x - y||^2), where x is the input and y is the user-provided "correct" values.

### Random forests for image approximation

```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

#### a) Get Mona Lisa Image

```python
mona = imageio.imread('mona_lisa.jpg')
```

```python
plt.imshow(mona)
```

#### b) Preprocessing of Input

```python
# check shape
mona.shape
```

```python
coord_rand_set = set()
coord_rand = []

# prevent doubles
while len(coord_rand_set) < 5000:
    x_rand = (np.random.randint(0,899))
    y_rand = (np.random.randint(0,603))
    if (x_rand, y_rand) not in coord_rand_set:
        coord_rand_set.add((x_rand,y_rand))
        coord_rand.append((x_rand, y_rand))

# check correct length
(np.asarray(coord_rand)).shape
```

To preprocess the data, we can normalize the image by dividing it by 255. This would allow the dot products to be 
done more efficiently and prevent the data from keeping increasing with the products. Since it is also needed for imshow to plot later, we can normalize at this point and forget about it. Other preprocessing steps include filling missing data values, or converting categorical values to numerical values, but these are not necessary in this case.

```python
# normalize it
mona = mona/255
```

#### c) Preprocessing of Output

```python
from skimage.color import rgb2gray
gray_mona = rgb2gray(mona)
```

```python
plt.imshow(gray_mona, cmap = 'gray')
plt.savefig('Mona_Grey.png')
```

Since the necessary information in the image is available in grayscale, we can just perform the model on the grayscale image. This automatically gives us normalized values for the image (i.e. pixel values 0 to 1 instead of 0 to 255).

#### d) Feed coordinates to random forest and color pixels with result

```python
y_training = [] # to store the colored pixels - esentially the labels
mona_coord = []

for i in range(900):
    for j in range(604):
        mona_coord.append((i,j))

for i in range(5000):
    y_training.append(list(mona[coord_rand[i]]))

# set up regressor  
regressor = RandomForestRegressor(max_depth=20) 

# fit regressor 
regressor.fit(coord_rand, y_training) 

# predict
predicted_lisa = regressor.predict(mona_coord)

# resshape
lisa_reshape = predicted_lisa.reshape(900,604, 3)

# display predicted image
plt.imshow(lisa_reshape)
plt.savefig('Lisa_regression.png')
```

#### e) Experimentation

##### i) Repeat the experiment for a random forest containing a single decision tree with depths 1, 2, 3, 5, 10, and 15. Explain the results.

```python
# function to apply regressor with different depths
def applyRegressor(depth):
    regressor = RandomForestRegressor(max_depth = depth, n_estimators=1)
    regressor.fit(coord_rand, y_training)
    predicted_lisa = regressor.predict(mona_coord)
    lisa_reshape = predicted_lisa.reshape(900,604, 3)
    return lisa_reshape
```

```python
depths = [1, 2, 3, 5, 10, 15]

for i, depth in enumerate(depths):
    lisa = applyRegressor(depths[i])
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust(right=2, top=2)
    plt.imshow(lisa)
    
plt.savefig('Result_e_i.png')
```

As the depth increases, the quality of the image reconstruction increases. For small changes in depth, there is a big change in quality. This is because as we increase the depth of the tree the number of weak classifiers that we are using increases. Increasing the number of weak classifers gives us a stronger classifier.

The image clarity slightly increases as the number of trees increases and then begins to blur. The change is not very noticeable. This is because with the increase in the number of trees you are not increasing the complexity of the classifier within one tree. The complexity of the classifier within one tree remains the same as before. When you increase the number of trees, the different trees might have slight differences in their classifiers and when all of the trees are combined, this is what we see in the images above. Thus, when there are fewer trees the images look more discrete because there is less variation because of the fact that there are fewer trees. When there is a medium number of trees you have the most discrete image because the variation in the different trees may be helping us to get more of the features represented. Finally, when there are too many trees it blurs because of the averaging of variation in many trees.

##### iv) Pruning Strategies

Some ways to prune the decision tree is to change the maximum depth (which we already tried),
changing the number of maximum leaf nodes, and changing the size of the sample leaf

```python
# function to apply regressor with different number of maximum leaf nodes
def applyRegressorLeaf(leaf_nodes):
    regressor = RandomForestRegressor(max_depth = 20, max_leaf_nodes = leaf_nodes)
    regressor.fit(coord_rand, y_training)
    predicted_lisa = regressor.predict(mona_coord)
    lisa_reshape = predicted_lisa.reshape(900,604, 3)
    return lisa_reshape
```

```python
leaf_nodes = [10, 20, 30, 50, 100, 500]

for i, depth in enumerate(leaf_nodes):
    lisa = applyRegressor(leaf_nodes[i])
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust(right=2, top=2)
    plt.imshow(lisa)
    
plt.savefig('Pruning_1.png')
```

We can see from above that increasing the maximum number of leaf nodes leads to better accuracy of the image. This is
because the more leafs the smaller the image sections are, and the more granular the reconstruction is.

```python
# function to apply regressor with different minimum size of samples leaf
def applyRegressorLeaf(size):
    regressor = RandomForestRegressor(max_depth = 20, min_samples_leaf = size)
    regressor.fit(coord_rand, y_training)
    predicted_lisa = regressor.predict(mona_coord)
    lisa_reshape = predicted_lisa.reshape(900,604, 3)
    return lisa_reshape
```

```python
size = [30, 100, 300] # try allowed values according to library

for i, depth in enumerate(size):
    lisa = applyRegressor(size[i])
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust(right=2, top=2)
    plt.imshow(lisa)
    
plt.savefig('Pruning_2.png')
```

##### (ii) Why the figure looks the way it does?

The resulting image looks like there are square patches of color of different sizes that are arranged in almost a circular way. The reason it looks like this is because the decision tree is made of a bunch of weak classifiers. Each of these weak classifiers creates a split in the decision tree and therefore a split in the image. This is why the patches of color are square and are arranged in this way.

##### (iii)

The number of square patches of color in the resulting image if there is only one tree in the forest is 2^d, where d is the depth of the tree. 

##### (iv)

If the forest contains n decision trees then there are n trees of depth d. This means that the resulting image for each of these trees independently is 2^d. However, in a random forest we are combining the tree estimators. Thus, it is possible for there to be n(2^d) square patches of color in the resulting image. However, this may not be obvious because the differences in the trees may be so subtle that many will overlap and leave us with fewer, perhaps more blurred patches of color. The patches of color that we are able to see may actually be closer to 2^d.
